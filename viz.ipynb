```python
import matplotlib.pyplot as plt
```

```python
with open('log.txt', 'r') as f:
    log = f.read()

log = log.split('\n')
log = [l.split(',') for l in log][:-1]
y = []
for l in log:
    y.append(float(l[0]))
```

```python
import torch
y_ = torch.tensor(y)
segments = y_.unfold(0, size=10, step=1)
mid = segments.mean(dim=-1)
std = segments.std(dim=-1)
epochs = range(mid.size(0))

plt.plot(epochs, mid)
plt.fill_between(x=epochs, y1=mid+std, y2=mid-std, alpha=0.2)
plt.show()
```
